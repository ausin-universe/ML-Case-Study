## Decision Tree
Decision Trees (DTs) are a non-parametric supervised learning method used for classification and regression.<br> The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.

```python
from sklearn import tree
```

```python
X = [[1,2],[3,4]]
X
```

```python
y = [1,2]
```

```python
a = np.array(X)
a
```

```python
b = np.array(y)
b
```

```python
clf = tree.DecisionTreeClassifier()
clf
```

```python
clf = clf.fit(a, b)
clf
```

```python
clf.predict([[2., 2.]])
```

```python
tree.plot_tree(clf)
```
