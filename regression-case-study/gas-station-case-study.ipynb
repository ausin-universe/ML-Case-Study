```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score,mean_squared_error,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
electric = pd.read_csv('datasets/gas_station.csv')
electric
```

```python
model = LinearRegression()
model
```

```python
X = electric.drop(['el_power'],axis=1)
X
```

```python
y = electric['el_power']
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=42)
```

```python
X_train
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
visz = pd.DataFrame(data=y_pred,columns=['Predicted'])
visz
```
