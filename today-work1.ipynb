```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
```

```python
data = {
    'X':[1,1.5,3,5,3.5,4.5],
    'Y':[1,1.5,3,8,5,5.5]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['X'],df['Y'],color='blue')
for i in range(df.shape[0]):
  plt.text(df['X'][i]+0.1, df['Y'][i], f"Point {i+1}")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Data Points')
plt.show()
```

```python
linked = linkage(df, method='ward')
```

```python
plt.figure(figsize=(8,6))
dendrogram(linked, labels=[f'Point {i+1}' for i in range(df.shape[0])])
plt.title('Dendrogram for small datasets')
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')
lables = cluster.fit_predict(df)
```

```python
df['Cluster'] = lables
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['X'], df['Y'], c=df['Cluster'], cmap='viridis')
for i in range(df.shape[0]):
  plt.text(df['X'][i]+0.1, df['Y'][i], f"Point {i+1}")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Hierarchical Clustering of Small Dataset')
plt.show()
```
